```python
import pypdfium2 as pdfium
from pathlib import Path

import os, sys
sys.path.append(os.path.abspath('../'))
from src.helper import prepend_zero
```

```python
ROOT_DIR = Path().resolve().parent
```

```python
pdf = pdfium.PdfDocument("../datasets/statistik-indonesia-2024-combined.pdf")
for i, page in enumerate(pdf):
    bitmap = page.render(
        scale = 1/72 * 150, # 150dpi
    )

    pil_image = bitmap.to_pil()
    pil_image.save(f"{ROOT_DIR.joinpath('outputs', 'pdfium_150dpi')}/{prepend_zero(str(i+1))}")
    print(f"Page {i+1} successfuly rendered")
```
