# Membuat Dataset Persentase CMYK pada Setiap Gambar dengan Berbagai DPI

```python
import os
import sys
import time

sys.path.append(os.path.abspath('..'))
from src.helper import log, rgb_to_cmyk, calculate_cmyk_percentage
```

```python
dpi_list = [150, 100, 50, 20, 10, 5]

for dpi in dpi_list:
    for index, img_file_name in enumerate(os.listdir(f"../outputs/pdfium_{dpi}dpi/")):
        start = time.time()
        if os.path.isfile(f"../outputs/pdfium_{dpi}dpi/{img_file_name}"):
            c,m,y,k = calculate_cmyk_percentage(f"../outputs/pdfium_{dpi}dpi/{img_file_name}")
            processing_time = time.time() - start
    
            # page,dpi,processing_time,c,m,y,k
            log('../datasets/cmyk-by-dpi.csv', f"{index+1},{dpi},{processing_time},{c},{m},{y},{k}")
            print(f"{index+1},{dpi},{processing_time},{c},{m},{y},{k}")
            del start, processing_time
```
